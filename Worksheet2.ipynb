```python
df = pd.read_csv('/content/drive/MyDrive/Dataset/medical_students_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
null_cols = df.columns[df.isnull().any()]
null_cols
```

```python
numeric_cols = ['Age', 'Height', 'Weight', 'BMI', 'Temperature', 'Heart Rate', 'Blood Pressure', 'Cholesterol']
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())
```

```python
categorical_cols = ['Gender', 'Blood Type', 'Diabetes', 'Smoking']
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
titanic_df=pd.read_csv('/content/drive/MyDrive/Dataset/Titanic-Dataset.csv')
```

```python
titanic_df.head()
```

```python
titanic_df.info()
```

```python
subset_df = titanic_df[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']]

first_class_df = subset_df[subset_df['Pclass'] == 1]

print("Mean Fare:", first_class_df['Fare'].mean())
print("Median Fare:", first_class_df['Fare'].median())
print("Max Fare:", first_class_df['Fare'].max())
print("Min Fare:", first_class_df['Fare'].min())
```

```python
print("Null values in Age:", first_class_df['Age'].isnull().sum())
first_class = first_class_df.dropna(subset=['Age'])
first_class.head()
```

```python
dummies = pd.get_dummies(titanic_df['Embarked'], prefix='Embarked')
titanic_df= pd.concat([titanic_df, dummies], axis=1)
titanic_df = titanic_df.drop('Embarked', axis=1)
titanic_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

print(titanic_df.groupby('Sex')['Survived'].mean())
sns.barplot(x='Sex', y='Survived', data=titanic_df)
plt.title("Survival Rate by Gender")
plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', hue='Embarked_C', data=titanic_df)
plt.title("Survival Rate by Gender & Cherbourg Embarkation")
plt.show()
```

```python
titanic_df['Age_group'] = pd.qcut(titanic_df['Age'], 5)
survival = titanic_df.groupby(['Pclass', 'Age_group'])['Survived'].mean()

sns.heatmap(survival.unstack(), annot=True, cmap="YlGnBu")
plt.title("Survival Rate by Age Group & Passenger Class")
plt.show()
```
